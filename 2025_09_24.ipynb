```python
import numpy as np
import pandas as pd

%precision 3
```

```python
df = pd.read_csv('../notebook/data/ch2_scores_em.csv',
                 index_col = 'student number')
```

```python
en_scores = np.array(df['english'])[:10]
ma_scores = np.array(df['mathematics'])[:10]

scores_df = pd.DataFrame({'english' : en_scores,
                          'mathematics' : ma_scores},
                         index = pd.Index(['A', 'B', 'C', 'D', 'E',
                                           'F', 'G', 'H', 'I', 'J'],
                                          name = 'student'))
scores_df
```

```python
summary_df = scores_df.copy()
# 각 변량 - 평균 = 편차
summary_df['english_deviation'] = summary_df['english'] - summary_df['english'].mean()
summary_df['mathematics_deviation'] = summary_df['mathematics'] - summary_df['mathematics'].mean()
# (X편차 * Y편차)의 평균 = 공분산 
summary_df['product_of_deviations'] = summary_df['english_deviation'] * summary_df['mathematics_deviation']

summary_df
```

```python
# 공분산
summary_df['product_of_deviations'].mean()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
english_scores = np.array(df['english'])
math_scores = np.array(df['mathematics'])

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111)

ax.scatter(english_scores, math_scores)
ax.set_xlabel('english')
ax.set_ylabel('mathematics')

plt.show()
```

```python
# 1차 다항식으로 회귀 (y = θ0 + θ1 * x)
poly_fit = np.polyfit(english_scores, math_scores, 1)  # 계수 [θ1, θ0]
poly_1d = np.poly1d(poly_fit)                         # 다항식 객체

# 직선을 그리기 위한 x 좌표 생성
xs = np.linspace(english_scores.min(), english_scores.max())

# 대응하는 y 좌표 계산
ys = poly_1d(xs)

# 시각화
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.set_xlabel('english')
ax.set_ylabel('mathematics')

# 산점도
ax.scatter(english_scores, math_scores, label='score')

# 회귀직선
ax.plot(xs, ys, color='gray',
        label=f'{poly_fit[1]:.2f} + {poly_fit[0]:.2f}x')

# 범례
ax.legend(loc='upper left')

plt.show()
```

```python
fig = plt.figure(figsize = (10, 8))
ax = fig.add_subplot(111)

c = ax.hist2d(english_scores, math_scores,
              bins = [9, 8], range = [(35, 80), (55, 95)])
ax.set_xlabel('english')
ax.set_ylabel('mathematics')
ax.set_xticks(c[1])
ax.set_yticks(c[2])

fig.colorbar(c[3], ax = ax)
plt.show()
```

```python
anscombe_data = np.load('../notebook/data/ch3_anscombe.npy')
print(anscombe_data.shape)
anscombe_data[0]
```

```python
stats_df = pd.DataFrame(index = ['X.mean', 'X_variance', 'Y_mean',
                                 'Y_variance', 'X&Y_corrlation',
                                 'X&Y_regression line'])

for i, data in enumerate(anscombe_data) :
    dataX = data [:, 0]
    dataY = data [:, 1]
    poly_fit = np.polyfit(dataX, dataY, 1)
    stats_df[f'data{i + 1}'] = [f'{np.mean(dataX):.2f}', f'{np.var(dataX):.2f}', f'{np.mean(dataY):.2f}', f'{np.var(dataY):.2f}',
                                f'{np.corrcoef(dataX, dataY)[0, 1]:.2f}', f'{poly_fit[1]:.2f} + {poly_fit[0]:.2f}x']
stats_df
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 10),
                         sharex = True, sharey = True)

xs = np.linspace(0, 30, 100)
for i, data in enumerate(anscombe_data) :
    poly_fit = np.polyfit(data[:,0], data[:,1], 1)
    poly_1d = np.poly1d(poly_fit)
    ys = poly_1d(xs)

    ax = axes[i // 2, i % 2]
    ax.set_xlim([4, 20])
    ax.set_ylim([3, 13])

    ax.set_title(f'data{i + 1}')
    ax.scatter(data[:,0], data[:,1])
    ax.plot(xs, ys, color = 'gray')

plt.tight_layout()
plt.show()
```
